```python
# Suma de dos número enteros

print("--------------------------------")
print("------Suma dos Enteros ---------")
print("--------------------------------")

#input
x = int(input("Digite el valor de X: "))
y = int(input("Digite el valor de Y: "))

#procesing

z = x + y

#output

print(" La suma de ", x, " mas ", y, " es igual a ",z)
```

```python
# Potencia de un número

print("--------------------------------")
print("------potencia de x ------------")
print("--------------------------------")

#input
x = int(input("Digite el valor de X: "))
y = int(input("Digite el valor de Y: "))

#procesing

z = x**y

#output

print(x, " Elevado a la ", y, " es ",z)
```
